**Set Example**

```python
 
Days=set(["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
Months={"Jan","Feb","Mar"}
Dates={21,22,17}
print(Days)
print(Months)
print(Dates)
```

**Print Set Elements **

```python
 
Days=set(["Mon","Tue","Wed","Thu","Fri","Sat","Sun"])
 
for day in Days:
	print(day)
```

**Add Element to set**

```python

Days=set(["Mon","Tue","Wed","Thu","Fri","Sat"])
 
Days.add("Sun")
print(Days)
```

**Remove element**

```python
 
Days=set(["Mon","Tue","Wed","Thu","Fri","Sat"])
 
Days.discard("Sun")
print(Days)
```
