**Import Dequeue**

```python
import collections
```

```python
DoubleEnded = collections.deque(["January","February","March"])
DoubleEnded.append("April")
print(DoubleEnded)
```

**Append on left**

```python
DoubleEnded.appendleft("May")
print ("Appended at left is - ")
print (DoubleEnded)
```

**Delete on right**

```python
DoubleEnded.pop()
print ("Deleting from right - ")
print (DoubleEnded)
```

**Delete on left**

```python
DoubleEnded.popleft()
print ("Deleting from left - ")
print (DoubleEnded)
```
