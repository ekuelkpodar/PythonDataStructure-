**Importing Heap**

```python
import heapq
```

**Inserting data into heap**

```python
heap = [31,11,48,85,35,9]

heapq.heapify(heap)
print(heap)
```

**Add Item **

```python
heapq.heappush(heap,8)
print(heap)
```

**Remove Item**

```python
heapq.heappop(heap)

print(heap)
```

**Replace Item**

```python
heapq.heapify(heap)
print(heap)

# Replace an element
heapq.heapreplace(heap,6)
print(heap)
```
