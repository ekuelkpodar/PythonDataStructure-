```python
import collections

dict1 = {'Month1': 'January', 'Month': 'February'}
dict2 = {'Month3': 'March', 'Month1': 'April'}

res = collections.ChainMap(dict1, dict2)

print(res.maps,'\n')

print('Keys = {}'.format(list(res.keys())))
print('Values = {}'.format(list(res.values())))
print()

print('elements:')
for key, val in res.items():
    print('{} = {}'.format(key, val))
print()

print('Month3 in res: {}'.format(('Month1' in res)))
print('Month4 in res: {}'.format(('Month4' in res)))

 
```

```python


res1 = collections.ChainMap(dict1, dict2)

print(res1.maps,'\n')

res2 = collections.ChainMap(dict2, dict1)

print(res2.maps,'\n')
 
```

```python


res = collections.ChainMap(dict1, dict2)

print(res.maps,'\n')

dict2['Month4'] = 'April'

print(res.maps,'\n')
```
