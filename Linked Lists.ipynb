**Creating a LinkedList**

```python

class Node:
  # init function
  def __init__(self, data = None, next=None): 
    self.data = data
    self.next = next


class LinkedList:
  def __init__(self):  
    self.head = None
  
  def insert(self, data):
    newNode = Node(data)
    if(self.head):
      current = self.head
      while(current.next):
        current = current.next
      current.next = newNode
    else:
      self.head = newNode
  
  def print(self):
    current = self.head
    while(current):
      print(current.data)
      current = current.next

```

**Create a linkedList and Print its elements**

```python
llist = LinkedList()
llist.insert(10)
llist.insert(20)
llist.insert(30)
llist.insert(40)
llist.insert(50)
llist.print()
```
