**Binary Search**

```python
def bsearch(list, val):
   size = len(list) - 1
   left = 0
   right = size
# Get the middle most value
   while left <= right:
      mid = (left + right)// 2
      if list[mid] == val:
         return mid
# Compare current the value with middle value
   if val > list[mid]:
      left = mid + 1
   else:
      right = mid - 1
   if left > right:
      return None
# Initialize the sorted list
list = [3,8,18,37,56,78]

# Print the search result
print(bsearch(list,73))
print(bsearch(list,13))
```
