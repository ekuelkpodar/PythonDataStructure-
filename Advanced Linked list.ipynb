```python
class Node:
   def __init__(self, data):
      self.data = data
      self.next = None
      self.prev = None
class doubly_linked_list:
   def __init__(self):
      self.head = None

   def push(self, NewVal):
      NewNode = Node(NewVal)
      NewNode.next = self.head
      if self.head is not None:
         self.head.prev = NewNode
      self.head = NewNode

   def append(self, NewVal):
      NewNode = Node(NewVal)
      NewNode.next = None
      if self.head is None:
         NewNode.prev = None
         self.head = NewNode
         return
      last = self.head
      while (last.next is not None):
         last = last.next
      last.next = NewNode
      NewNode.prev = last
      return

   def listprint(self, node):
      while (node is not None):
         print(node.data),
         last = node
         node = node.next

dl = doubly_linked_list()
dl.push(15)
dl.append(6)
dl.push(7)
dl.push(45)
dl.append(89)
dl.listprint(dl.head)
```
