** Declare a dictionary **

```python
dict = {'Name': 'John', 'Age': 27, 'Class': 'College'}
```

**Getting Data with key**

```python
print("dict['Name']: ", dict['Name'])
print("dict['Age']: ", dict['Age'])
```

**Updating Key**

```python
dict['School'] = "GSU"; 
```

```python
print("['Age']: ", dict['Age'])
print("['School']: ", dict['School'])
```
